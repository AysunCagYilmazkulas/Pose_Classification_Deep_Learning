```python
# Reading CSV file.

df = pd.read_csv('goodmorning_csvs_out.csv', names=column_names, header=None)
df.head(1000)
```

```python
data = df.drop(df.filter(regex='z').columns, axis=1)
data.head(1000)
```

```python
# We can change the categorical values to integer.

data["cat"]=[0 if i.strip()=="Down" else 1 for i in data.cat]
# data["cat"]=[0 if i==1 else 1 if i==2 else 3 for i in data.cat]
data.head(1000)
```

```python
col_names = ['image_name', 'cat', 'torso_left_shoulder_right_shoulder', 'left_shoulder-left_wrist(11-15)'
, 'right_shoulder-right_elbow(12-14)', 'right_shoulder-right_ankle(12-28)'
,'right_elbow-right_wrist(14-16)', 'left_wrist-right_wrist(15-16)'
, 'left_wrist-left_hip(15-23)', 'left_wrist-left_ankle(15-27)'
, 'left_hip-right_hip(23-24)', 'left_hip-left_ankle(23-27)'
, 'right_hip-right_knee(24-26)', 'right_knee-right_ankle(26-28)'
,'left_ankle-right_ankle(27-28)', 'left_hip-left_knee(23-25)', 'left_knee-left_ankle(25-27)'
, 'left_shoulder-left_elbow(11-13)', 'left_elbow-left_wrist(13-15)'
, 'right_hip-right_ankle(24-28)', 'right_shoulder-right_wrist(12-16)']
```

```python
def distance(x1,y1,x2,y2):
    return ((x1-x2)**2+(y1-y2)**2)**0.5

#distance(828.98895, 235.87022, 1009.50757, 242.48903)
# a=(828.98895-1009.50757)
# b=(235.87022-242.48903)
# c=(a**2+b**2)**0.5
# print(c)
```

```python
image_names=[]
categories=[]
torso=[]
left_shoulder_left_wrist=[]
right_shoulder_right_elbow=[]
right_shoulder_right_ankle=[]
right_elbow_right_wrist=[]
left_wrist_right_wrist=[]
left_wrist_left_hip=[]
left_wrist_left_ankle=[]
left_hip_right_hip=[]
left_hip_left_ankle=[]
right_hip_right_knee=[]
right_knee_right_ankle=[]
left_ankle_right_ankle=[]
left_hip_left_knee=[]
left_knee_left_ankle=[]
left_shoulder_left_elbow=[]
left_elbow_left_wrist=[]
right_hip_right_ankle=[]
right_shoulder_right_wrist=[]
```

```python
for i in range(len(data)):
    image_names.append(data.iloc[i]['image_name'])
    categories.append(data.iloc[i]['cat'])
    torso.append(distance(data['x11=left_shoulder'][i],data['y11'][i], data['x12=right_shoulder'][i], data['y12'][i]))
    left_shoulder_left_wrist.append(distance(data['x11=left_shoulder'][i],data['y11'][i], data['x15=left_wrist'][i], data['y15'][i]))
    right_shoulder_right_elbow.append(distance(data['x12=right_shoulder'][i], data['y12'][i],data['x14=right_elbow'][i], data['y14'][i]))
    right_shoulder_right_ankle.append(distance(data['x12=right_shoulder'][i],data['y12'][i],data['x28=right_ankle'][i],data['y28'][i]))
    right_elbow_right_wrist.append(distance(data['x14=right_elbow'][i],data['y14'][i],data['x16=right_wrist'][i],data['y16'][i]))
    left_wrist_right_wrist.append(distance(data['x15=left_wrist'][i],data['y15'][i],data['x16=right_wrist'][i], data['y16'][i],))
    left_wrist_left_hip.append(distance(data['x15=left_wrist'][i],data['y15'][i],data['x23=left_hip'][i], data['y23'][i]))
    left_wrist_left_ankle.append(distance(data['x15=left_wrist'][i],data['y15'][i],data['x27=left_ankle'][i], data['y27'][i]))
    left_hip_right_hip.append(distance(data['x23=left_hip'][i],data['y23'][i],data['x24=right_hip'][i],data['y24'][i]))
    left_hip_left_ankle.append(distance(data['x23=left_hip'][i],data['y23'][i],data['x27=left_ankle'][i],data['y27'][i]))
    right_hip_right_knee.append(distance(data['x24=right_hip'][i],data['y24'][i],data['x26=right_knee'][i],data['y26'][i]))
    right_knee_right_ankle.append(distance(data['x26=right_knee'][i],data['y26'][i], data['x28=right_ankle'][i],data['y28'][i]))
    left_ankle_right_ankle.append(distance(data['x27=left_ankle'][i],data['y27'][i],data['x28=right_ankle'][i],data['y28'][i]))
    left_hip_left_knee.append(distance(data['x23=left_hip'][i],data['y23'][i],data['x25=left_knee'][i],data['y25'][i]))
    left_knee_left_ankle.append(distance(data['x25=left_knee'][i],data['y25'][i],data['x27=left_ankle'][i],data['y27'][i]))
    left_shoulder_left_elbow.append(distance(data['x11=left_shoulder'][i],data['y11'][i], data['x13=left_elbow'][i],data['y13'][i]))
    left_elbow_left_wrist.append(distance(data['x13=left_elbow'][i],data['y13'][i],data['x15=left_wrist'][i],data['y15'][i]))
    right_hip_right_ankle.append(distance(data['x24=right_hip'][i],data['y24'][i], data['x28=right_ankle'][i],data['y28'][i]))
    right_shoulder_right_wrist.append(distance(data['x12=right_shoulder'][i],data['y12'][i],data['x16=right_wrist'][i],data['y16'][i]))
```
